```python
import numpy as np
import math
from matplotlib import pyplot as plt
import statsmodels.api as sm
from scipy import stats

%matplotlib inline
```

```python
def make_ar1(rho, n):
    epsilon = np.random.randn(n)
    z = [epsilon[0]]
    for i, eps in enumerate(epsilon[1:]):
        a = rho * z[i] + eps
        z.append(a)
    return z

y = make_ar1(rho = 5, n = 20)
```

```python
### a.

rho = .5
n = 1000
y = make_ar1(rho, n)

y_1 = y[0:-1]
X = np.column_stack((np.ones(len(y)-1), y[1:]))

fit1 = sm.OLS(y_1, X).fit()

fig, ax = plt.subplots(figsize = (14, 5))
ax.set_title('AR(1) process with rho = %2.2f and n = %d' %(rho, n))
ax.plot(y, lw = 0.9)
fig.savefig('problem1a.png', bbox_inches='tight')
plt.show()


print fit1.summary()
```

###### Problem 2b.

```python
rho = 1
n = 1000
slopes = []
for i in range(1000):
    w1l = make_ar1(rho, n)
    w2l = make_ar1(rho, n)
    Xl = np.column_stack((np.ones(len(w2l)), w2l))
    slope = sm.OLS(w1l, Xl).fit().params[1]
    slopes.append(slope)
slopes = np.array(slopes)
```

```python
from matplotlib import rc
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
x = np.linspace(-3, 3)
fig, ax = plt.subplots(figsize = (10, 5))
ax.hist(slopes, lw = 1, normed = 1, color = "darkcyan", label = r'\textrm{regression coefficients} \beta_1') #label = "Regression coefficients")
ax.plot(x, stats.norm.pdf(x), '--', color = "darkorange", lw = 2, label = "Standard normal")
ax.plot(x,
        stats.norm.pdf(x, loc = slopes.mean(), scale = slopes.std()),
        '-',
        color = "darkorange",
        lw = 2,
        label = ("Normal(%2.2f, %2.2f)" %(round(slopes.mean(),2)+0, slopes.std()**2)))
ax.set_title(r"Normalized histogram of regression coefficients $w_2 = \beta_0 + \beta_1w_1$")
ax.legend(loc = 1)
plt.savefig('problem2b.png', bbox_inches = 'tight')
plt.show()
```
