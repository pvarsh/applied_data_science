```python
import patsy
### sys, os, utilities
import os
import sys
import csv
import timeit
from datetime import datetime
### scikit-learn
from sklearn.linear_model import LinearRegression
### statsmodels
import statsmodels.api as sm
### numpy, scipy, matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


%matplotlib inline
plt.style.use('ggplot')
```

###### Read data

```python
def readData_noParse():
    fileIn = "../data/train.csv"
    train = pd.read_csv(fileIn, index_col = ['datetime'])
    columns = list(train.columns)
    columns[-1] = 'count_tot'
    train.columns = columns
    return train

train = readData_noParse()
train.head()
print "Index of type: ", type(train.index[0])
train.index = pd.DatetimeIndex(train.index)
print "Index of type: ", type(train.index[0])
seasonsDict = {1: 'spring', 2: 'summer', 3: 'fall', 4: 'winter'}
train['seasons'] = train.season.apply(lambda s: seasons[s])
train.head(3)
```

###### Resample in hours to get NA in missing data

```python
train_rs = train.resample('H', how = sum) # rs stands for resampled
train_rs['ind'] = np.array(range(train_rs.shape[0])) + 1
train_rs.head(2)
```

### Detrend
* resample by date using sum
* fit linear model
* divide slope by 24

```python
### aggregate by month
train_month = pd.DataFrame(train_rs.count_tot.resample('M', how = sum))
train_month['ind'] = np.array(range(train_month.shape[0])) + 1 # no real reason for + 1

### plot
fig, [ax1, ax2] = plt.subplots(nrows = 2, figsize = (20, 5))
ax1.plot(train_rs.count_tot, label = "Hourly data")
### fit and plot hourly data
y, X = patsy.dmatrices('count_tot ~ ind', data = train_rs, return_type = 'dataframe')
model = sm.OLS(y, X).fit()
beta0, beta1 = model.params
print "Hourly linear fit: intercept = %2.2f; slope = %2.4f" %(beta0, beta1)
ax1.plot(train_rs.ind, beta0 + beta1 * train_rs.ind, label = "Linear fit to hourly data")

### fit daily data and plot
y, X = patsy.dmatrices('count_tot ~ ind', data = train_month, return_type = 'dataframe')
model = sm.OLS(y, X).fit()
beta0, beta1 = model.params
print "Monthly linear fit: intercept = %2.2f; slope = %2.4f; hour_slope = %2.4f" %(beta0, beta1, beta1/(24*30))
ax2.plot(train_month.count_tot, label = "Daily data")
ax2.plot(model.predict(X))

ax.legend(loc = 'best')
plt.show()
```
