```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use('ggplot')
```

```python
X = [[6], [8], [10], [14],   [18]]
y = [[7], [9], [13], [17.5], [18]]
model = LinearRegression()
model.fit(X, y)
model.predict(X) - y
```

```python
fig, axs = plt.subplots(ncols = 2, figsize = (12, 5))
axs[0].plot(X, y, label = "Data")
axs[0].plot(X, model.predict(X), label = "Regression line")
axs[1].plot(y - model.predict(X), label = "Residuals", marker = 'o', lw = 0)
plt.show()
```
